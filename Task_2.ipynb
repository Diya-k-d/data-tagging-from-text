```python
import pandas as pd

df = pd.read_excel("/DA -Task 2..xlsx")
df.head()
```

```python
# Drop completely empty rows
df = df.dropna(how='all')

# Standardize text columns
for col in df.select_dtypes(include='object').columns:
    df[col] = df[col].astype(str).str.strip().str.title()

# Remove duplicates
df = df.drop_duplicates()

df.isnull().sum()
```

```python
df.columns
```

```python
text_cols = ['CUSTOMER_VERBATIM', 'CORRECTION_VERBATIM', 'CAUSAL_PART_NM', 'GLOBAL_LABOR_CODE_DESCRIPTION']

for col in text_cols:
    df[col] = df[col].astype(str).str.strip()
```

```python
df['CAUSAL_PART_NM'] = df['CAUSAL_PART_NM'].fillna("Unknown")
df['CUSTOMER_VERBATIM'] = df['CUSTOMER_VERBATIM'].fillna("Not Provided")
df['CORRECTION_VERBATIM'] = df['CORRECTION_VERBATIM'].fillna("Not Provided")
df['GLOBAL_LABOR_CODE_DESCRIPTION'] = df['GLOBAL_LABOR_CODE_DESCRIPTION'].fillna("Unknown")
```

```python
df['TOTALCOST'] = pd.to_numeric(df['TOTALCOST'], errors='coerce')
```

```python
df[['CAUSAL_PART_NM','GLOBAL_LABOR_CODE_DESCRIPTION','TOTALCOST']].head()
```

```python
import matplotlib.pyplot as plt

df['CAUSAL_PART_NM'].value_counts().head(10).plot(kind='bar')
plt.title("Top 10 Causal Parts by Frequency")
plt.xlabel("Causal Part")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['GLOBAL_LABOR_CODE_DESCRIPTION'].value_counts().head(10).plot(kind='bar')
plt.title("Top Repair Categories")
plt.xlabel("Repair Category")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df.groupby('CAUSAL_PART_NM')['TOTALCOST'].sum().sort_values(ascending=False).head(10).plot(kind='bar')
plt.title("Top 10 Parts by Total Cost Impact")
plt.xlabel("Causal Part")
plt.ylabel("Total Cost")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
df['GLOBAL_LABOR_CODE_DESCRIPTION'].value_counts().head(10).plot(kind='bar')
```

```python
df.groupby('CAUSAL_PART_NM')['TOTALCOST'].sum().plot(kind='bar')
```

```python
df.to_excel("Task2_Cleaned_Tagged.xlsx", index=False)
```

```python
df['CAUSAL_PART_NM'] = df['CAUSAL_PART_NM'].replace("Nan", "Unknown")
df['CAUSAL_PART_NM'] = df['CAUSAL_PART_NM'].fillna("Unknown")
```

```python
def generate_tags(text):
    text = str(text).lower()
    tags = []
    if "leak" in text:
        tags.append("Leak")
    if "noise" in text:
        tags.append("Noise Issue")
    if "sensor" in text:
        tags.append("Sensor Issue")
    if "overheat" in text:
        tags.append("Overheating")
    if "vibration" in text:
        tags.append("Vibration")
    if "stall" in text:
        tags.append("Stalling")
    if "brake" in text:
        tags.append("Brake Issue")
    if "transmission" in text:
        tags.append("Transmission Issue")
    if "steering" in text:
        tags.append("Steering Issue")
    if "wheel" in text:
        tags.append("Wheel Issue")
    return ", ".join(tags)

df['Generated_Tags'] = df['CUSTOMER_VERBATIM'].apply(generate_tags)
```

```python
df.to_excel("Task2_Cleaned_Tagged.xlsx", index=False)
```
